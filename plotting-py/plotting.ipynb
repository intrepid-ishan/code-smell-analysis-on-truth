```python
import plotly.io as pio
pio.renderers.default = 'iframe'
```

```python
%matplotlib inline
```

```python
import pandas as pd
```

```python
df_result = pd.DataFrame(columns = ['Commit_No','Smells_Count','LOC','Design_Smell_Density','Implementation_Smell_Density', 'Architecture_Smell_Density','LOC_max','LOC_avg','LCOM_max','LCOM_avg','WMC_max','WMC_avg'])
for i in range(1,11):
    
    df_design = pd.read_csv(str(i)+"/DesignSmells.csv",error_bad_lines=False)
    design_count = df_design.shape[0]
    
    df_Implement = pd.read_csv(str(i)+"/ImplementationSmells.csv",error_bad_lines=False)
    implement_count = df_Implement.shape[0]
    
    df_arch = pd.read_csv(str(i)+"/ArchitectureSmells.csv",error_bad_lines=False)
    arch_count = df_arch.shape[0]
    
    df_metrics = pd.read_csv(str(i)+"/TypeMetrics.csv",error_bad_lines=False)
    
    loc_sum = df_metrics['LOC'].sum()
    loc_max = df_metrics['LOC'].max()
    loc_avg = df_metrics['LOC'].mean()
    
    df_metrics['LCOM'] = df_metrics['LCOM'].replace(-1,0)
    df_metrics['WMC'] = df_metrics['WMC'].replace(-1,0)
    
    lcom_sum = df_metrics['LCOM'].sum()
    lcom_max = df_metrics['LCOM'].max()
    lcom_avg = df_metrics['LCOM'].mean()
    
    wmc_sum = df_metrics['WMC'].sum()
    wmc_max = df_metrics['WMC'].max()
    wmc_avg = df_metrics['WMC'].mean()
    df_result = df_result.append({'Commit_No':i, 
                                  'Smells_Count':implement_count+design_count+arch_count, 
                                  'LOC':loc_sum, 
                                  'Design_Smell_Density':(design_count*1000)/loc_sum, 
                                  'Implementation_Smell_Density':(implement_count*1000)/loc_sum,
                                  'Architecture_Smell_Density':(arch_count*1000)/loc_sum,
                                  'LOC_max':loc_max,
                                  'LOC_avg':loc_avg,'LCOM_max':lcom_max,
                                  'LCOM_avg':lcom_avg,
                                  'WMC_max':wmc_max,
                                  'WMC_avg':wmc_avg},ignore_index=True)
#     print(design_count,implement_count,loc_sum)
```

```python
import plotly.express as px

# df = px.data.gapminder().query("country=='Canada'")
fig = px.line(df_result, x=df_result['Commit_No'], y=df_result['Smells_Count'], title='Smell Count vs Commit No.',color_discrete_sequence=['#0077c2', '#f48c06'])
fig.update_layout(
    plot_bgcolor='#E7ECEF'
)
fig.show()
```

```python
fig = px.line(df_result, x=df_result['Commit_No'], y=df_result['LOC'], title='LOC vs Commit No.',color_discrete_sequence=['#0077c2', '#f48c06'])
fig.update_layout(
    plot_bgcolor='#E7ECEF'
)
fig.show()
```

```python
fig = px.line(df_result, x=df_result['Commit_No'], y=df_result['Design_Smell_Density'], title='Design Smell Density vs Commit No.', color_discrete_sequence=['#0077c2', '#f48c06'])
fig.update_layout(
    plot_bgcolor='#E7ECEF'
)
fig.show()
```

```python
fig = px.line(df_result, x=df_result['Commit_No'], y=df_result['Implementation_Smell_Density'], title='Implementation Smell Density vs Commit No.', color_discrete_sequence=['#0077c2', '#f48c06'])
fig.update_layout(
    plot_bgcolor='#E7ECEF'
)
fig.show()
```

```python
fig = px.line(df_result, x=df_result['Commit_No'], y=df_result['Architecture_Smell_Density'], title='Architecture Smell Density vs Commit No.', color_discrete_sequence=['#0077c2', '#f48c06'])
fig.update_layout(
    plot_bgcolor='#E7ECEF'
)
fig.show()
```

```python
import plotly.graph_objects as go
fig = go.Figure()
fig.add_scatter(x=df_result['Commit_No'], y=df_result['LOC_max'], mode='lines',name = 'LOC_max',showlegend=True,line=dict(color='#0077c2'))
fig.add_scatter(x=df_result['Commit_No'], y=df_result['LOC_avg'], mode='lines',name = 'LOC_avg',showlegend=True,line=dict(color='#f48c06'))
fig.update_layout(
    xaxis_title="Commits",
    yaxis_title="Line Count",
    title="Commit vs Max and Avg LOC(Line of Code)",
    plot_bgcolor='#E7ECEF'
)
fig.show()
```

```python
fig = go.Figure()
fig.add_scatter(x=df_result['Commit_No'], y=df_result['LCOM_max'], mode='lines',name = 'LCOM_max',showlegend=True, line=dict(color='#0077c2'))
fig.add_scatter(x=df_result['Commit_No'], y=df_result['LCOM_avg'], mode='lines',name = 'LCOM_avg',showlegend=True, line=dict(color='#f48c06'))
fig.update_layout(
    xaxis_title="Commits",
    yaxis_title="LCOM",
    title="Commit vs Max and Avg LCOM(Lack of Cohesion of Methods)",
    plot_bgcolor='#E7ECEF'
)
fig.show()
```

```python
fig = go.Figure()
fig.add_scatter(x=df_result['Commit_No'], y=df_result['WMC_max'], mode='lines',name = 'WMC_max',showlegend=True, line=dict(color='#0077c2'))
fig.add_scatter(x=df_result['Commit_No'], y=df_result['WMC_avg'], mode='lines',name = 'WMC_avg',showlegend=True, line=dict(color='#f48c06'))
fig.update_layout(
    xaxis_title="Commits",
    yaxis_title="WMC",
    title="Commit vs Max and Avg WMC(Weighted Methods per Class)",
    plot_bgcolor='#E7ECEF'
)
fig.show()
```
